```python
# Membuat fitur gabungan untuk Content-Based Filtering
anime["features"] = (
    anime["Genres"] + " " +
    anime["Type"] + " " +
    anime["Studios"] + " " +
    anime["Rating"]
)
```

```python
# Melihat hasil fitur gabungan
anime[["Name", "features"]].head()
```

```python
# Mengecek tipe data fitur numerik
anime[["Episodes", "Score", "Popularity"]].dtypes
```

Menggabungkan beberapa atribut anime menjadi satu fitur agar dapat diproses menggunakan TF-IDF.

```python
# Membuat objek TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
```

```python
# Mengubah fitur teks menjadi matriks TF-IDF
tfidf_matrix = tfidf.fit_transform(anime["features"])
```

```python
# Melihat ukuran matriks TF-IDF
tfidf_matrix.shape
```

```python
# Melihat beberapa kata yang dipelajari TF-IDF
tfidf.get_feature_names_out()[:20]
```

```python
# Menghitung kemiripan antar anime
cosine_sim = cosine_similarity(tfidf_matrix)
```

```python
# Melihat ukuran matriks kemiripan
cosine_sim.shape
```

```python
# Membuat model KNN
knn = NearestNeighbors(
    n_neighbors=10,
    metric='cosine'
)
```

```python
# Melatih model KNN
knn.fit(tfidf_matrix)
```

```python


# Ubah dari float64 ke float32
cosine_sim = cosine_sim.astype(add_np.float32)

# Folder penyimpanan
save_path = "model"
os.makedirs(save_path, exist_ok=True)

# Simpan TF-IDF
with open(os.path.join(save_path, "tfidf.pkl"), "wb") as file:
    pickle.dump(tfidf, file)

# Simpan cosine similarity
with open(os.path.join(save_path, "cosine_similarity.pkl"), "wb") as file:
    pickle.dump(cosine_sim, file, protocol=pickle.HIGHEST_PROTOCOL)

print(f"File berhasil disimpan di: {save_path}")
```

```python
# Menyimpan model KNN
with open("knn_model.pkl", "wb") as file:
    pickle.dump(knn, file)
```

```python
anime = anime[
    [
        "Name",
        "Genres",
        "Type",
        "Episodes",
        "Studios",
        "Duration",
        "Score",
        "Popularity",
        "Rating"
    ]
]

# Simpan dataset hasil preprocessing
anime.to_csv("anime_clean.csv", index=False)

print("anime_clean.csv berhasil disimpan")
```
